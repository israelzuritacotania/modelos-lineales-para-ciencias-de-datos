## Análisis estadístico y visual de las variables seleccionadas
A continuación veremos el análisis estadístico y visual de cada variable, agrupando para cada una: el código, la gráfica y la interpretación, tomando de base el codigo visto en clases e interpretando segun los resultados.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Función para graficar tasa de aprobación por variable categórica
def plot_approval_rate(df, var, target='y'):
    rates = df.groupby(var)[target].mean().reset_index()
    plt.figure(figsize=(6,4))
    # Paleta personalizada: naranja y azul
    custom_palette = ['#FFA500', '#1E90FF', '#FFA500', '#1E90FF', '#FFA500', '#1E90FF', '#FFA500', '#1E90FF']
    n_cats = rates[var].nunique()
    palette = custom_palette[:n_cats] if n_cats <= len(custom_palette) else None
    sns.barplot(x=var, y=target, data=rates, palette=palette)
    plt.title(f'Tasa de aprobación según {var}')
    plt.ylabel('Tasa de aprobación')
    plt.xlabel(var)
    plt.ylim(0,1)
    plt.show()
```

```python
# address
plot_approval_rate(df_Xy, 'address')
functions.ml_eda_functions.ttest_with_shapiro(df_Xy, 'address', 'y', clear_x='no', intervals=2, lower_y=0, apply_trans='no', equals_part='yes')
```

**Interpretación:**
La tasa de aprobación suele ser mayor en estudiantes con residencia urbana ('U'), posiblemente esto por el mejor acceso a recursos educativos. Sin embargo, es una variable no gestionable directamente por la escuela.

## Hipótesis y acciones para variables gestionables

A continuación se proponen hipótesis y acciones para las variables gestionables o parcialmente gestionables identificadas:

### 1. Acceso a internet en casa (internet)
- **Hipótesis:** Mejorar el acceso a internet en casa incrementa la tasa de aprobación.
- **Acción:** Implementar programas municipales o escolares para dotar de acceso a internet a estudiantes vulnerables (becas de conectividad, convenios con proveedores, espacios de estudio con wifi en la escuela).

### 2. Tiempo de viaje a la escuela (traveltime)
- **Hipótesis:** Reducir el tiempo de traslado mejora el rendimiento académico.
- **Acción:** Gestionar rutas de transporte escolar, promover el uso de bicicletas dependiendo de la ciudad que se estudio, o flexibilizar horarios para quienes viven lejos.

### 3. Tutor principal (guardian)
- **Hipótesis:** El acompañamiento parental directo favorece la aprobación.
- **Acción:** Programas de orientación a familias para fortalecer el rol parental y la comunicación escuela-hogar, especialmente en casos donde el tutor no es madre/padre.


## Análisis extra: tasa de aprobación cruzada

## Análisis extra: SQL sobre el dataset

A continuación veremos un ejemplo de consulta SQL usando pandasql para analizar la tasa de aprobación según acceso a internet.

```python
# Análisis SQL con Spark SQL en Databricks
# Convertir a tabla temporal
spark_df = spark.createDataFrame(df_Xy)
spark_df.createOrReplaceTempView('students')

# Consulta SQL: tasa de aprobación por colegio y acceso a internet
result = spark.sql('''
SELECT school, internet, AVG(y) as tasa_aprobacion
FROM students
GROUP BY school, internet
''')
display(result)
```

**Interpretación:**
La consulta SQL confirma que el acceso a internet está asociado a una mayor tasa de aprobación.
